# Audio Features Correlation
This notebook loads the audio features from the dataset and calculates the correlation matrix.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to the dataset
file_path = "/Users/gin/Documents/Visual Analytics/Project/dataset/final/masterlist_lyrics_with_features_cleaned_top300.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Define the audio feature columns (common Spotify audio features)
audio_features = [
    'acousticness', 'danceability', 'energy', 'instrumentalness', 
    'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'valence', 'duration_ms'
]

# Filter features that exist in the dataframe
available_features = [f for f in audio_features if f in df.columns]

# Check the head of the selected features
df[available_features].head()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[available_features].corr()

# Display the correlation matrix
correlation_matrix
```

```python
# Generate a heatmap for the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Audio Features')
plt.tight_layout()
plt.show()
```
